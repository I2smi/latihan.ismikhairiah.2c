```python
import matplotlib.pyplot as plt

# Data contoh
bulan = ['B1', 'B2', 'B3', 'B4', 'B5']
jumlah_pengeluaran = [80, 60, 55, 30, 20]

# grafiknya 
plt.figure(figsize=(8, 4))
plt.plot(bulan, jumlah_pengeluaran, marker='o', color='yellow', linestyle='-') 
plt.title('Line Chart Penjualan per bulan')
plt.xlabel('bulan')
plt.ylabel('Jumlah Penjualan')
plt.grid(True)
plt.show()
```

penjelasan :data diatas merupakan data jumlah penjualan perbulan menggunakan library matplotlib dengan bentuk grafik

```python
import matplotlib.pyplot as plt

# Data Contoh
tahun = ['T1', 'T2', 'T3', 'T4', 'T5', 'T6','T7']
jumlah_kerugian = [10, 30, 70, 50, 80, 70, 50]

# grafiknya
plt.figure(figsize=(8, 4))
plt.bar(tahun, jumlah_kerugian, color='red')
plt.title('Bar Chart Pengeluaran per tahun')
plt.xlabel('tahun')
plt.ylabel('Jumlah kerugian')
plt.grid(False)
plt.show()
```

penjelasan :data diatas merupakan data jumlah pengeluaran pertahun menggunakan library matplotlib dengan bentuk grafik batang
